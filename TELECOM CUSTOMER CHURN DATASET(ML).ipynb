```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings(action='ignore')
```

```python
df=pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\Telecom_Customer_Churn_Dataset.csv")
df.head()
```

### REMOVING ALL DUPLICATE AND NULL VALUES

```python
df=df.drop_duplicates()
df=df.dropna()
```

### FEATURE SCALING

```python
a=df.select_dtypes(include=np.number)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
a=StandardScaler().fit_transform(a)
```

```python
b=pd.DataFrame(a,columns=['Tenure_Months','Monthly_Charges','Total_Charges'])
```

```python
b.head()
```

### ENCODING

```python
c=df.select_dtypes(include=['object'])
```

```python
d=c.drop('CustomerID',axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in d:
    d[i]=LabelEncoder().fit_transform(d[i])
```

```python
d.head()
```

### COMBINING THE NUMERICAL AND CATEGORICAL COLUMNS

```python
df1=pd.concat([b,d],axis=1)
```

```python
df1.head()
```

### DEFINING THE TARGET AND FEATURE

```python
y=df1['Churn']
x=df1.drop(columns=['Churn'])
```

```python
x.head()
```

### TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

### MODEL DEPLOYMENT

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

### PREDICTED VALUES

```python
y_pred=model.predict(x_test)
```

### METRICES

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
